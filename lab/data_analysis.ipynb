```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_folder = "D:/Uni/GAN_powered_Pokemon_Generation/data/All_Generations"
```

```python
image_sizes = []
image_modes = []
image_aspect_ratios = []
```

```python
for file_name in os.listdir(image_folder):
    if file_name.endswith(".jpg"):
        file_path = os.path.join(image_folder, file_name)
        
        # Open image
        with Image.open(file_path) as img:
            # Save image size and  aspect ratio
            image_sizes.append(img.size)
            image_modes.append(img.mode)
            aspect_ratio = img.size[0] / img.size[1]  # Breite / HÃ¶he
            image_aspect_ratios.append(aspect_ratio)

image_sizes = np.array(image_sizes)
image_aspect_ratios = np.array(image_aspect_ratios)
```

```python
print("Minimum image size (Width, Height):", np.min(image_sizes, axis=0))
print("Maximum image size (Width, Height):", np.max(image_sizes, axis=0))
print("Average image size (Width, Height):", np.mean(image_sizes, axis=0))
```

```python
unique_modes = set(image_modes)
print("Unique color modes:", unique_modes)
```

```python
print("Average aspect ratio:", np.mean(image_aspect_ratios))
```

```python
plt.figure(figsize=(10, 5))
plt.hist(image_sizes[:, 0], bins=20, alpha=0.5, label='Width')
plt.hist(image_sizes[:, 1], bins=20, alpha=0.5, label='Height')
plt.title('Distribution of Image Sizes')
plt.xlabel('Pixels')
plt.ylabel('Number of Images')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(image_aspect_ratios, bins=20, color='orange', alpha=0.7)
plt.title('Distribution of Aspect Ratios')
plt.xlabel('Aspect Ratio (Width/Height)')
plt.ylabel('Number of Images')
plt.show()
```
